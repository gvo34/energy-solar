```python
## Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import os
```

## Source EIA

```python
## Source: https://www.eia.gov/totalenergy/data/browser/?tbl=T10.05#/?f=M
filename = "MER_T10_05_cut.csv"
filepath = os.path.join("data",filename)

eiadf = pd.read_csv(filepath)
eiadf.head()
```

```python
eiadf['Description'].value_counts()
```

```python
# values = eiadf[['Value', 'YYYYMM']]
# values = values.loc[values['Value'] != "Not Available"
values = eiadf[['YYYYMM','Value']]
values
```

```python
values.dtypes
```

```python
values.columns
```

```python
from datetime import datetime

listdf=[]
for i, row in values.iterrows():
    listdf.append(row)
ls1 = [i["Value"] for i in listdf]
ls2 = [datetime.strptime(str(int(i["YYYYMM"])),"%Y%m") for i in listdf]
#ls2 = [str(int(i["YYYYMM"])) for i in listdf]
#print(ls2)
newdf = pd.DataFrame(data={"YYYYMM":ls2,"Value":ls1, })
newdf.dtypes
```

```python
newdf.to_csv("data/monthdatacut.csv", index=False)
```

```python
#remove values totals respresented with YYYY13

newdf.plot()
plt.show()
```
